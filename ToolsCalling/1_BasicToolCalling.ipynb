```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
#tool create

@tool
def multiply(a:int, b:int) ->int:
    """Multiply these two numbers"""
    return a*b
```

```python
print(multiply.invoke({'a':3,'b':5}))
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

# Tool Binding

```python
llm=ChatOpenAI()
```

```python
llm.invoke('hi')
```

```python
llm_with_tool = llm.bind_tools([multiply])
```

```python
llm_with_tool.invoke('hi')
```

```python
llm_with_tool.invoke('multiply 3 with 20')
```

```python
llm_with_tool.invoke('multiply 3 with 20').tool_calls[0]
```
