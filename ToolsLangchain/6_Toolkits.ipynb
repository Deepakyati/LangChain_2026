# Toolkit

```python
#reusability
#multiple tools are available
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add(a:int, b:int) ->int:
    """Add two numbers"""
    return a+b

@tool
def multiply(a:int, b:int) ->int:
    """multiply two numbers"""
    return a*b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add,multiply]
```

```python
toolkit=MathToolkit()

tools=toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```
