# Custom Tools

# @Tool decorators

```python
from langchain_core.tools import tool
```

```python
#create a function
#add type hints
#add tool decorators


@tool
def multiply(a:int,b:int) ->int:
    """Multiply two numbers"""
    return a*b
```

```python
result=multiply.invoke({"a":3, "b":6})

print(result)
```

```python
#tool name
#tool description
#tool arguments

print(multiply.name)
print(multiply.description)
print(multiply.args)
```
