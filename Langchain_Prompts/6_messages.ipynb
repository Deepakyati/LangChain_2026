```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

#HumanMessage - user messages send to llm
#AIMessage - llm response
#SystemMessage - like - you are a helpful assistant, you are a doctor etc.

load_dotenv()

model = ChatOpenAI()
```

```python
#HumanMessage - user messages send to llm
#AIMessage - llm response
#SystemMessage - like - you are a helpful assistant, you are a doctor etc.

messages=[
    SystemMessage(content="You are a helpful assistant that translates English to French."),
    HumanMessage(content="Translate the following English text to French: 'Hello, how are you?'")
]
```

```python

result = model.invoke(messages)

messages.append(AIMessage(content=result.content))

print(messages)
```
