TypedDict 

way to define Dictionary in Python, where you specify what keys and values should exist.
- helps to ensure dict structure
- Cons: it doesn't validate 

```python
"""Example TypedDict code:

from typing_extensions import TypedDict
from typing import TypedDict

class Person(TypedDict):
    name: str
    age: int
    email: str

    new_person: Person = {
    "name": "Alice",
    "age": 30,
    "email": "alice@example.com" 
}"""
```

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict

load_dotenv()

model=ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)
```

```python
#schema
class Review(TypedDict):
    summary: str
    sentiment: str
```

```python
structured_output=model.with_structured_output(Review)
```

```python
#product review
result = structured_output.invoke("Lenovo is a great laptop brand. Their laptops are known for their durability and performance. I recently bought a Lenovo ThinkPad and it has exceeded my expectations. The keyboard is comfortable to type on, the battery life is impressive, and the overall build quality is top-notch. I would highly recommend Lenovo laptops to anyone in need of a reliable and efficient device.")
```

```python
print(result)
print(result['summary'])
print(result['sentiment'])
```
